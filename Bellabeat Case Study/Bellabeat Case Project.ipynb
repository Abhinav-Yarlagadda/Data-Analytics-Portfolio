### Merging and Performing Operations on Datasets 

```python
new_daily_activity = pd.DataFrame()
df = daily_activity.groupby(['Id'], as_index=False)["Calories"].mean()
df2 = daily_activity.groupby(['Id'], as_index=False)["ActivityDate"].count()
df3 = daily_activity.groupby(['Id'], as_index=False)["TotalSteps"].mean()
df4 = daily_activity.groupby(['Id'], as_index=False)["TotalDistance"].mean()
df5 = daily_activity.groupby(['Id'], as_index=False)["LightActiveDistance"].mean()
df6 = daily_activity.groupby(['Id'], as_index=False)["ModeratelyActiveDistance"].mean()
df7 = daily_activity.groupby(['Id'], as_index=False)["VeryActiveDistance"].mean()
new_daily_activity["Id"] = df["Id"]
new_daily_activity["Number of Days"] = df2["ActivityDate"]
new_daily_activity["Mean Calories"] = df["Calories"]
new_daily_activity["Mean Steps"] = df3["TotalSteps"]
new_daily_activity["Mean Distance"] = df4["TotalDistance"]
new_daily_activity["Mean Light Intensity"] = df5["LightActiveDistance"]
new_daily_activity["Mean Moderate Intensity"] = df6["ModeratelyActiveDistance"]
new_daily_activity["Mean High Intensity"] = df7["VeryActiveDistance"]
display(new_daily_activity)
```

## Data Analysis and Graphing 

```python
import seaborn as sns

#Creating Scatterplot
x = new_daily_activity["Mean Steps"]
y = new_daily_activity["Mean Calories"]
plt1 = plt.scatter(new_daily_activity["Mean Steps"], new_daily_activity["Mean Calories"], s=20)
plt.title("Mean Calories Burned vs Mean Steps", fontdict={"fontsize": 12})
plt.xlabel("Mean Steps")
plt.ylabel("Mean Calories")

#Creating Line of Regression with Confidence Intervals
sns.regplot(x=x, y=y)
display(plt1)

#Finding Correlation Coefficient
r = np.corrcoef(x, y)
print(f"r = {r[0, 1]}")
```

### Analysis 

* The scatterplot above confirms the positive relationship between calories burned and steps taken, albeit the relationship being relatively weak with an r-correlation of 0.436
* In addition, the confidence intervals appear to be wide, which reveals that simply taking more steps doesn't always translate to buring more calories
    * However, the data does imply the health benefits of taking more steps as it usually means more calories are burned as it keeps the users active and increasing their fitness, which tremendously benefits their well-being in return

```python
x = new_daily_activity["Mean Distance"]
y = new_daily_activity["Mean Calories"]
plt2 = plt.scatter(new_daily_activity["Mean Distance"], new_daily_activity["Mean Calories"], s=20)
plt.title("Mean Calories Burned vs Mean Distance", fontdict={"fontsize": 12})
plt.xlabel("Mean Distance")
plt.ylabel("Mean Calories")
sns.regplot(x=x, y=y)
display(plt2)
r = np.corrcoef(x, y)
print(f"r = {r[0, 1]}")
```

### Analysis 

* The data above demonstrates a moderately positive relationship between calories burned and distance, which is stronger than the relationship with calories burned and steps
* The difference in correlation strength of distance and steps suggests that distance may serve as the better measurement of activity level for fitness goals
    * One possible reason for this difference could be that not all FitBit users travel the same distance with each of their steps, which explains the different regression models
    * Even so, using distance to measure fitness still comes with limitations as the confidence intervals are wide, so simply traveling a longer distance does not guarantee that more calories will be burned

```python
x = new_daily_activity["Mean Light Intensity"]
y = new_daily_activity["Mean Calories"]
plt3 = plt.scatter(new_daily_activity["Mean Light Intensity"], new_daily_activity["Mean Calories"], s=20)
plt.title("Mean Calories Burned vs Mean Light Intensity", fontdict={"fontsize": 12})
plt.xlabel("Mean Light Intensity")
plt.ylabel("Mean Calories")
sns.regplot(x=x, y=y)
display(plt3)
r = np.corrcoef(x, y)
print(f"r = {r[0, 1]}")
```

### Analysis 

* The relationship above between calories burned and the light intensity level is depicted as a weak, but positive correlation
    * Although there exists a slightly positive correlation between activities done with light intensity and calories burned, it is not meaningful enough to justify that more calories will be burned by doing light exercise
    * One rationale for why a positive relationship may exist at all could be on the account of people who do more light activity are more likely to engage in more intense exercise as well, instead of remaining sedentary

```python
x = new_daily_activity["Mean Moderate Intensity"]
y = new_daily_activity["Mean Calories"]
plt4 = plt.scatter(new_daily_activity["Mean Moderate Intensity"], new_daily_activity["Mean Calories"], s=20)
plt.title("Mean Calories Burned vs Mean Moderate Intensity", fontdict={"fontsize": 12})
plt.xlabel("Mean Moderate Intensity")
plt.ylabel("Mean Calories")
sns.regplot(x=x, y=y)
display(plt4)
r = np.corrcoef(x, y)
print(f"r = {r[0, 1]}")
```

### Analysis 

* As the scatterplot demonstrates above there appears to be no correlation between calories burned and a moderate intensity level
    * One possible explaination for this scenario would be due to the limited amount of time spent doing activities involving what is to be considered as a moderate intensity level
    * Another possible explanation is that when FitBit users were trying to exercise, they spent most of their time in the moderate intensity level, meaning less time spent doing high intensity activities, thereby lowering the amount of calories burnt

```python
x = new_daily_activity["Mean High Intensity"]
y = new_daily_activity["Mean Calories"]
plt5 = plt.scatter(new_daily_activity["Mean High Intensity"], new_daily_activity["Mean Calories"], s=20)
plt.title("Mean Calories Burned vs Mean High Intensity", fontdict={"fontsize": 12})
plt.xlabel("Mean High Intensity")
plt.ylabel("Mean Calories")
sns.regplot(x=x, y=y)
display(plt5)
r = np.corrcoef(x, y)
print(f"r = {r[0, 1]}")
```

### Analysis 

* The scatterplot reveals that there exists a moderately positive correlation between calories burned and a high intensity level
* Therefore, the data verifies that high intensity exercise is more conducive to burning calories
* On the other hand, the data can't be used to prove causation as a person doesn't always burn extra calories just by doing more high intensity exercise
    * For example, a person may do a desk job (which is mostly sedentary) and perform high intensity exercise for one hour, but a person who is continously walking throughout the day doing a mix of light and moderate intensity activity could still burn more calories than them

```python
x = sleep_day["TotalMinutesAsleep"]
y = sleep_day["TotalTimeInBed"]
plt6 = plt.scatter(sleep_day["TotalMinutesAsleep"], sleep_day["TotalTimeInBed"], s=10)
plt.title("Total Time in Bed vs Total Minutes Asleep", fontdict={"fontsize": 12})
plt.xlabel("Total Minutes Asleep")
plt.ylabel("Total Time in Bed")
sns.regplot(x=x, y=y)
display(plt6)
r = np.corrcoef(x, y)
print(f"r = {r[0, 1]}")
```

### Analysis 

* The scatterplot above demonstrates an extremely strong positive correlation between time alseep and time in bed
* More importantly, there are a few outliers which aren't on the ling of regression, which indicates users potentially suffering from sleep disorders
    * Therefore, the data collected from these smart devices may be useful in the future for aiding diagnosis in sleep disorders like insomnia

```python
x = weight_log["WeightKg"]
y = weight_log["BMI"]
plt7 = plt.scatter(weight_log["WeightKg"], weight_log["BMI"], s=5)
plt.title("BMI vs Weight (kg)", fontdict={"fontsize": 12})
plt.xlabel("Weight (kg)")
plt.ylabel("BMI")
sns.regplot(x=x, y=y)
display(plt7)
r = np.corrcoef(x, y)
print(f"r = {r[0, 1]}")
```

### Analysis 

* As the scatterplot depicts, there exists a strong positive correlation between BMI and weight
    * While the data portrays weight as generally increasing in relation to BMI, it is also important to note that people who weigh more do not always have a higher BMI than others
        * Some possible reasons which may account for this difference may include height, age, and gender

* It is also worthwhile to mention that not all users provided a weight log to analyze data from, which means this scatterplot is limited in the sense that it uses insufficient data samples to generalize the results
    * Consequently, while the data collected is from a wide spectrum, most of the users were concentrated on the lower end of the spectrum
    * As the outlier represents, smart devices can be used to combat obsesity (BMI > 30) and while not displayed in the data, these device may help also identify those underweight (BMI < 18)

```python
#Calculating Totals of Columns for Slices
light_intensity = new_daily_activity["Mean Light Intensity"].sum()
moderate_intensity = new_daily_activity["Mean Moderate Intensity"].sum()
high_intensity = new_daily_activity["Mean High Intensity"].sum()

#Creating Pie Chart
slices = [light_intensity, moderate_intensity, high_intensity]
labels = ["Light Intensity", "Moderate Intensity", "High Intensity"]
explode = [0, 0, 0.06]
colors = ["tomato", "palegreen", "deepskyblue"]
plt.style.use("default")
plt.pie(slices, labels = labels, 
        colors = colors, wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = "%1.1f%%")
plt.title("Intensity Type for Total Distance")
plt.tight_layout()
plt.show()
```

### Analysis 

* Each slice refers to the overall average intensity of activity done for the total distance traveled by FitBit users to compare the general activity level of its users
    * The largest slice is the light intensity portion of the total distance, which mainly entails just walking for day to day activities
    * The high intensity slice was emphasized in the pie chart to demonstrate the importance of maintaining fitness to smart device users
    * Surprisingly, the moderate intensity slice was the smallest portion, which is likely due to most users moving with a purpose either for just walking or for exercise, which is tends to be high energy
